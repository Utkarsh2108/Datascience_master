### `Q1`. What is Flask Framework? What are the advantages of Flask Framework?


- Flask is a lightweight and flexible web framework for Python that allows you to build web applications quickly and easily. It is known for its simplicity, ease of use, and flexibility, and is particularly well-suited for building small to medium-sized web applications.

#### Advantages of Flask Framework:

1. Lightweight and easy to use: Flask is a lightweight framework that is easy to use and learn, making it a great choice for beginners.

2. Flexible and customizable: Flask is highly customizable and can be used for a variety of web development tasks. It provides a lot of flexibility and allows you to choose the components and features you need.

3. Extensible: Flask is an extensible framework that allows you to add new functionality using Flask extensions. There are many third-party extensions available for Flask that can be easily integrated into your application.

4. Minimalistic: Flask has a minimalistic approach that allows you to build your application from the ground up, giving you complete control over every aspect of the application.

5. Large Community: Flask has a large and active community of developers who contribute to its development and provide support to its users. This makes it easy to find solutions to common problems and get help when you need it.

6. Built-in Development Server: Flask comes with a built-in development server that makes it easy to test and debug your application.

### Q2. Create a simple Flask application to display ‘Hello World!!’. 
Attach the screenshot of the output in Jupyter Notebook.


```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello():
    return '<h2>Hello World!</h2>'
    
if __name__=="__main__":
    app.run(host="0.0.0.0")
```

### `Q3`. What is App routing in Flask? Why do we use app routes?


- In Flask, app routing refers to the process of mapping URLs to view functions or endpoints within a web application. It allows you to define how URLs should be handled and what response should be returned for each URL.
- App routing is the process of mapping URLs to view functions in a Flask web application. Flask provides a simple and flexible routing system that allows you to define routes for your application using the @app.route() decorator.

- The `@app.route()` decorator is used to associate a URL with a view function. It takes a URL pattern as an argument, and when a user requests that URL, Flask calls the associated view function to generate the response. 
- Here is an example of a simple app route:


```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello():
    return 'Hello, World!'
```

### `Q5`. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

- The url_for() function is used in Flask for URL building. It generates a URL to a specific function/view based on its name and any arguments that are passed to it.

```python
from flask import Flask
from flask import url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'index'

@app.route('/login')
def login():
    return 'login'

@app.route('/user/<username>')
def profile(username):
    return f'{username}\'s profile'

with app.test_request_context():
    print(url_for('index'))
    print(url_for('login'))
    print(url_for('login', next='/'))
    print(url_for('profile', username='Utkarsh'))
    
if __name__=="__main__":
    app.run(host="0.0.0.0")
```
