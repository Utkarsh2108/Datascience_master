2. **Median**: The median is the middle value in a dataset when it is arranged in order. If the number of observations is even, the median is the average of the two middle values. The median is less sensitive to extreme values than the mean and is often used when the data is skewed or has outliers.

### `Q5`. How are measures of dispersion such as range, variance, and standard deviation used to describe the spread of a dataset? Provide an example.


**Range** is the simplest measure of dispersion, and it represents the difference between the maximum and minimum values in a dataset. However, it only considers two values and does not take into account the entire spread of the data.

**Variance** and **Standard deviation** are more commonly used measures of dispersion. Variance measures the average distance of each observation from the mean of the dataset, while standard deviation is the square root of variance. A high variance or standard deviation indicates that the data is spread out over a larger range, while a low variance or standard deviation indicates that the data is clustered closely around the mean.

*  For example, consider a dataset that represents the number of hours of sleep per night for a group of people. The mean value is 7 hours, and the standard deviation is 1.5 hours. This means that on average, people in the group sleep for 7 hours per night, and the spread of sleep times around this mean is relatively large, with some people sleeping as little as 4.5 hours and others sleeping as much as 9 hours. In contrast, if the standard deviation were only 0.5 hours, this would indicate that the sleep times are tightly clustered around the mean, with most people sleeping for approximately 7 hours per night.

### `Q8`. What do you understand about skewness in data?


* Skewness is a measure of the degree of asymmetry in a distribution of data. In other words, it measures how lopsided a distribution is.

  * If a **distribution is symmetrical**, it has zero skewness. If the distribution is skewed to the right, it has positive skewness, and if it is skewed to the left, it has negative skewness.

  * In a **positively skewed distribution**, the tail of the distribution is longer on the right-hand side, and the median is typically lower than the mean. This is because the mean is pulled higher by the larger values on the right-hand side.

  *  In a **negatively skewed distribution**, the tail of the distribution is longer on the left-hand side, and the median is typically higher than the mean. This is because the mean is pulled lower by the smaller values on the left-hand side.

### `Q9`. If a data is right skewed then what will be the position of median with respect to mean?


*  If a data is right-skewed, the median will typically be less than the mean. This is because the skewed distribution is pulled towards the right by the presence of larger values in the right tail. As a result, the mean is shifted towards the right, while the median remains closer to the bulk of the data, which is towards the left.

### `Q10`. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?


Covariance and correlation are both measures of the relationship between two variables in statistical analysis. However, they differ in the way they quantify this relationship and the scale of their values.

**Covariance** : is a measure of the direction and strength of the linear relationship between two variables.
*  It measures how much two variables change together, with positive values indicating that they tend to move in the same direction, and negative values indicating that they tend to move in opposite directions. 
* The magnitude of the covariance depends on the units of measurement of the variables, making it difficult to compare covariances across different datasets.

**Correlation** : is a standardized measure of the linear relationship between two variables. 
*  It takes values between -1 and 1, with values of -1 indicating a perfect negative correlation (i.e., the variables move in opposite directions), values of 1 indicating a perfect positive correlation (i.e., the variables move in the same direction), and values of 0 indicating no correlation. 
*  Correlation coefficients are not affected by the scale of measurement of the variables, making them easier to compare across different datasets.

### `Q12`. For a normal distribution data what is the relationship between its measure of central tendency?


* For a normal distribution, the measures of central tendency (i.e., mean, median, and mode) are equal to each other. 
* This means that the data is symmetrically distributed around the central point, with half of the observations falling on each side of the mean.
* This is one of the defining characteristics of a normal distribution, and it allows for easy interpretation and analysis of the data. 
* the standard deviation of a normal distribution can be used to quantify the spread of the data around the mean

### `Q13`. How is covariance different from correlation?


1. **Scale**: 
*  Covariance values can range from negative infinity to positive infinity and depend on the scale of the variables being measured. 
*  Correlation values range from -1 to 1 and are standardized, meaning they are independent of the scale of the variables.

2. **Interpretation**: 
* Covariance measures the direction and magnitude of the linear relationship between two variables, while correlation measures the strength and direction of the linear relationship between two variables.

3. Symmetry: 
*  Covariance is not necessarily symmetrical, meaning the covariance between variable X and variable Y may be different than the covariance between variable Y and variable X. 
* Correlation, however, is always symmetrical, meaning the correlation between variable X and variable Y is the same as the correlation between variable Y and variable X.

### `Q14`. How do outliers affect measures of central tendency and dispersion? Provide an example

*  Outliers are data values that are very different from the rest of the data. They can affect measures of central tendency and dispersion, which are statistics that describe the center and spread of a dataset.

*  **Centrol Tendancy without outliers**

```python
data = [10,20,30,40,50]
print(f'Mean: {np.mean(data)}')
print(f'Median: {np.median(data)}')
```

 *  **Centrol Tendancy with outliers**

```python
data = [10,20,30,40,1000]
print(f'Mean: {np.mean(data)}')
print(f'Median: {np.median(data)}')
```

*  Note that the mean of the data that contains outlier is not an accurate representation of the mean of that data

*  **Measures of Dispersion Without Outliers**

```python
data = [10,20,30,40,50]
print(f'Variance: {np.var(data)}')
print(f'Standard deviation: {np.std(data)}')
```

*  Measures of Dispersion Without Outliers


```python
data = [10,20,30,40,1000]
print(f'Variance: {np.var(data)}')
print(f'Standard deviation: {np.std(data)}')
```
