### `Q1`. What is the Probability density function?


* A probability density function (PDF) is a function that describes the probability distribution of a continuous random variable. It gives the probability of the variable taking on any specific value within a certain range of values.

* The PDF is a non-negative function, and its integral over the entire range of the variable is equal to 1. The integral of the PDF between two values represents the probability that the variable takes on a value within that range.


### `Q2`. What are the types of Probability distribution?


1. Normal Distribution: Also known as Gaussian distribution, it is a continuous probability distribution that has a bell-shaped curve. Many natural phenomena and measurement errors can be modeled using the normal distribution.

2. Binomial Distribution: It is a discrete probability distribution that describes the number of successes in a fixed number of independent trials. Each trial has only two possible outcomes, success or failure.

3. Poisson Distribution: It is a discrete probability distribution that describes the number of events occurring in a fixed time interval or in a specified region of space. Poisson distribution is often used to model rare events or count data.

4. Uniform Distribution: It is a continuous probability distribution where all outcomes have an equal chance of occurring. It is often used in simulations and modeling to represent situations where each outcome is equally likely.


###`Q3`. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1,500,200)
```

```python
def pdf(x,mean,sd):
  p_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
  return p_density
```

```python
mean = np.mean(x)
sd = np.std(x)
```

```python
pdf = pdf(x,mean,sd)
plt.plot(x , pdf)
plt.xlabel('data')
plt.ylabel('Probability densiy ')
```

###`Q4`. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.


* The properties of the binomial distribution are:

> 1. The experiment consists of a fixed number of independent trials, denoted by n.
> 2. Each trial has only two possible outcomes, usually labeled as "success" and "failure".
> 3. The probability of success is the same for each trial and is denoted by p.
> 4. The random variable of interest is the number of successes, denoted by X.
> 5. The distribution of X is called the binomial distribution.

* Two examples of events where binomial distribution can be applied are:

* Flipping a coin: Suppose you flip a coin n times and you want to know the probability of getting k heads. Each flip of the coin is an independent trial with two possible outcomes (heads or tails), and the probability of getting heads is the same for each trial (assuming the coin is fair). This is an example of a binomial distribution with n trials and p = 0.5 (assuming the coin is fair).

*  Quality control: Suppose you inspect n items on a production line and you want to know the probability of finding k defective items. Each inspection is an independent trial with two possible outcomes (defective or non-defective), and the probability of finding a defective item is the same for each inspection. This is an example of a binomial distribution with n trials and some probability p of finding a defective item (which may depend on the production process).

###`Q6`. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.


```python
from math import exp, factorial

def poisson_cdf(mu, x):
    cdf = 0.0
    for i in range(x + 1):
        cdf += exp(-mu) * mu**i / factorial(i)
    return cdf
```

```python
cdf = poisson_cdf(mu=3,x=5)
print(cdf)
```

###`Q7`. How Binomial distribution different from Poisson distribution?


The key differences between the two distributions are:

1. Nature of the events: In the Binomial distribution, the events are discrete and have a binary outcome (success or failure), whereas in the Poisson distribution, the events are rare occurrences that can be considered continuous.

2. Number of trials/samples: The Binomial distribution models the number of successes in a fixed number of trials or samples, whereas the Poisson distribution models the number of rare events that occur in a fixed interval or sample size.

3. Probability of success/rate of occurrence: In the Binomial distribution, the probability of success is constant across trials or samples, whereas in the Poisson distribution, the rate of occurrence is constant over time or space.

4. Assumptions: The Binomial distribution assumes that the trials are independent and identically distributed, whereas the Poisson distribution assumes that the rare events occur independently of each other and that the rate of occurrence is constant over time or space.


###`Q8`. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.


```python
import numpy as np

np.random.seed(123)

sample = np.random.poisson(lam=5, size=1000)

sample_mean = np.mean(sample)
sample_variance = np.var(sample)

print("Sample mean:", sample_mean)
print("Sample variance:", sample_variance)
```

###`Q9`. How mean and variance are related in Binomial distribution and Poisson distribution?


1. Binomial distribution:
  * The mean of a binomial distribution with parameters n and p is μ = np.
  * The variance of a binomial distribution with parameters n and p is σ^2 = np(1-p)

2. Poisson distribution:
  * The mean of a Poisson distribution is μ = λ, where λ is the rate parameter.
  * The variance of a Poisson distribution is σ^2 = λ.

*  In both cases, the variance is a function of the mean.
*  For binomial distributions, the variance increases as p approaches 0 or 1, and is highest when p = 0.5.
*  For Poisson distributions, the variance is equal to the mean, indicating that the variability of the distribution is directly proportional to the mean.



###`Q10`. In normal distribution with respect to mean position, where does the least frequent data appear?

* In a normal distribution, the least frequent data appears in the tails of the distribution, which are the regions farthest away from the mean.

* For a normal distribution, approximately 68% of the data falls within one standard deviation of the mean, 95% falls within two standard deviations of the mean, and 99.7% falls within three standard deviations of the mean. The remaining data (0.3%) is distributed in the tails of the distribution, with half of it in the left tail and the other half in the right tail.

* Therefore, the least frequent data in a normal distribution will be found in the tails of the distribution, which are farthest away from the mean.
