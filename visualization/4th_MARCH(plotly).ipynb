### `Q1`: load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot ascatter plot for age and fare columns in the titanic dataset.


```python
import plotly.graph_objects as go
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=titanic.age, y=titanic.fare, mode='markers'))
fig.show()
```

### `Q2`:Using the tips dataset in the Plotly library, plot a box plot using Plotly express.


```python
tips= sns.load_dataset("tips")
tips.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Box(y=tips.total_bill))
fig.show()
```

### `Q3`:Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column inthe tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.


```python
import plotly.express as px
fig = px.bar(tips, x="sex", y="total_bill", color="day", pattern_shape="smoker")
fig.show()
```

### `Q5`: What is Distplot? Using Plotly express, plot a distplot.



* **Distplot:**
*  A Distplot or distribution plot, depicts the variation in the data distribution. 
* The distplot function in Python creates a histogram of the data, which shows the frequency of different values or ranges of values in the dataset.

```python
tips = sns.load_dataset("tips")
fig = px.histogram(tips, x="total_bill", y="tip", color="sex", marginal="rug",
                   hover_data=tips.columns)
fig.show()
```
