```python
import numpy as np
import matplotlib.pyplot as plt
```

### `Q1` What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

- Matplotlib is a popular data visualization library for Python. It provides a wide range of functions for creating static, animated, and interactive plots in various formats.

- Matplotlib is used to create data visualizations in fields such as science, engineering, finance, and data science. Some of the common use cases of Matplotlib include:
   - Exploratory data analysis
   - Data visualization for research papers and presentations  
   - Building dashboards and reports

#### 1. Line plot:
-  A line plot is a basic plot that shows the trend of a dataset over time.

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.figure(figsize=(3,3))
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line plot')
plt.show()
```

#### 2.Scatter plot:
- A scatter plot shows the relationship between two variables.

```python
plt.figure(figsize=(3,3))
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line plot')
plt.show()
```

#### 3.Bar plot:
- A bar plot is used to compare values across different categories

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [10, 7, 14, 8, 12]

plt.figure(figsize=(3,3))
plt.bar(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar plot')
plt.show()
```

#### 4.Histogram:
- A histogram is used to show the distribution of a dataset.

```python
x = np.random.normal(size=1000)

plt.figure(figsize=(3,3))
plt.hist(x, bins=30)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')
plt.show()
```

#### 5.Pie chart:
- A pie chart is used to show the proportion of different categories in a dataset.

```python
x = [30, 20, 15, 10, 5]
labels = ['A', 'B', 'C', 'D', 'E']

plt.figure(figsize=(3,3))
plt.pie(x, labels=labels)
plt.title('Pie chart')
plt.show()
```

### `Q3` Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
- For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
- For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
- For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
- For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

- The `subplot()` function in Matplotlib is used to create multiple plots within a single figure. It allows us to organize multiple plots in a grid-like structure, where each plot is placed in a separate cell of the grid.

```python
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig , axs= plt.subplots(2,2)

axs[0,0].plot(x1,y1)
axs[0,0].set_title('Line-1')

axs[0,1].plot(x2,y2)
axs[0,1].set_title('Line-2')

axs[1,0].plot(x3,y3)
axs[1,0].set_title('Line-3')

axs[1,1].plot(x4,y4)
axs[1,1].set_title('Line-4')

plt.subplots_adjust(hspace=0.5)

plt.show()
```

### `Q4` What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


- A bar plot is a type of data visualization that displays data using rectangular bars. The height or length of each bar represents the numerical value of a particular category or variable. Bar plots are commonly used to compare the values of different categories or to track changes in a variable over time.

- Bar plots are useful because they make it easy to visualize and compare the relative sizes of different categories or variables. They can also be used to highlight trends or patterns in the data.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(3,3))
plt.bar(company,profit)
plt.show()
```

```python
## horizontal bar plot.
plt.figure(figsize=(3,3))
plt.barh(company,profit)
plt.show()
```

### `Q5` What is a box plot? Why is it used? Using the following data plot a box plot.

- A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. 
- It shows the median (middle value), the quartiles (values that divide the data into four equal parts), and the range of the data. 
- The box represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3). 
- The whiskers extend from the box to show the range of the data, excluding any outliers. 
- Outliers are points that are more than 1.5 times the IQR away from the nearest quartile.

- Box plots are used to visually compare the distribution of data across different groups or to see the distribution of a single dataset. They can provide information about the central tendency, spread, and skewness of the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
x = [box1,box2]

fig, ax = plt.subplots()
ax.boxplot(x)

ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
ax.set_title('Box Plot')
plt.show()
```
